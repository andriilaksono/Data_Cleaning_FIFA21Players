### Cleaning Height and Weight Data
The height and weight columns initially contained non-numeric values ("cm" for height and "kg" for weight). To perform numerical analysis, we need to remove these units and convert the columns to numeric types.

### Height

```python
df['Height'].unique()
```

```python
# The height and weight columns contain feet/inches notation. These need to be cleaned and converted into numeric values for analysis
```

```python
import re

height = ['6\'2"', '6\'3"', '6\'5"', '5\'11"',
       '6\'4"', '6\'1"', '6\'0"', '5\'10"', '5\'9"', '5\'6"', '5\'7"',
       '5\'4"']


for h in height:
    match = re.match(r"(\d+)'(\d+)\"", h)
    if match:
        feet = int(match.group(1))
        inches = int(match.group(2))
        cm = feet * 30.48 + inches * 2.54
        cm_rounded = round(cm)  
        print(f"{h}: {cm_rounded} cm")
```

```python
df['Height'] = df['Height'].str.replace('6\'2"', "188cm")
df['Height'] = df['Height'].str.replace('6\'3"', "190cm")
df['Height'] = df['Height'].str.replace('6\'5"', "196cm")
df['Height'] = df['Height'].str.replace('5\'11"', "180cm")
df['Height'] = df['Height'].str.replace('6\'4"', "193cm")
df['Height'] = df['Height'].str.replace('6\'1"', "185cm")
df['Height'] = df['Height'].str.replace('6\'0"', "183cm")
df['Height'] = df['Height'].str.replace('5\'10"', "178cm")
df['Height'] = df['Height'].str.replace('5\'9"', "175cm")
df['Height'] = df['Height'].str.replace('5\'6"', "168cm")
df['Height'] = df['Height'].str.replace('5\'7"', "170cm")
df['Height'] = df['Height'].str.replace('5\'4"', "163cm")
```

```python
df['Height'].unique()
```

```python
df['Height'] = df['Height'].str.replace('cm', '')
df['Height'] = df['Height'].astype(int)
```

## Question 3
 Can you clean and transform the value, wage, and release clause columns into columns of integers?

```python
df[['Value', 'Wage', 'Release Clause']].head()
```

```python
df[['Value', 'Wage', 'Release Clause']].dtypes
```

```python
def convert_money(value):
    value = value.replace('€', '')
    if 'M' in value:
        return int(float(value.replace('M', '')) * 1e6)
    elif 'K' in value:
        return int(float(value.replace('K', '')) * 1e3)
    return int(float(value))

df['Value'] = df['Value'].apply(convert_money)
df['Wage'] = df['Wage'].apply(convert_money)
df['Release Clause'] = df['Release Clause'].apply(convert_money)
```

## Question 5
Should you separate the Team & Contract column into separate team and contract columns?

```python
df[['Club', 'Contract']]
```

```python
# The Club and Contract columns are already separated, but the Club column contains unnecessary newline characters, 
# and the Contract column includes extra whitespace. Both columns should be cleaned
```

```python


# Club
df['Club'] = df['Club'].str.replace('\n', '', regex=False)

# Contract
df[['Contract Start', 'Contract End']] = df['Contract'].str.split(' ~ ', expand=True)


df.drop('Contract', axis=1, inplace=True)

df.tail()
```
